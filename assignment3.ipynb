# 1. Understanding Pseudocode and Flowcharts #

Pseudocode is a way of representing an algorithm using a plain language-like structure that focuses on the logic of the program, without worrying about syntax. It is not a programming language, but an informal method to express algorithms in a readable way for humans.


A flowchart is a diagrammatic representation of a process or an algorithm. It uses different symbols to represent actions, decisions, and the flow of the process. Flowcharts help visualize the logic of a system step by step.

# How Pseudocode and Flowcharts Help in Algorithm Design #

Pseudocode: It allows the designer to focus on the logic rather than syntax, helping programmers understand the steps needed to solve a problem before writing the code.

Flowcharts: They help visualize the process of the algorithm. Flowcharts are ideal for understanding the sequence of steps, decisions, and loops, making it easier to debug and optimize the algorithm.

# 2. Finding the Largest of Three Numbers.
Pseudocode:


```python
Start
  Input num1
  Input num2
  Input num3
  If num1 >= num2 AND num1 >= num3
      Output num1 is the largest
  Else If num2 >= num1 AND num2 >= num3
      Output num2 is the largest
  Else
      Output num3 is the largest
End
```

Flowchart:

Start -> Input num1, num2, num3 -> Decision: Is num1 the largest? -> Yes -> Output num1 -> End
If No, Decision: Is num2 the largest? -> Yes -> Output num2 -> End
Else, Output num3 -> End.

# 5. Greatest Common Divisor (GCD) Calculation

Pseudocode:

```python
FUNCTION gcd(a, b)
  WHILE b != 0
    temp = b
    b = a MOD b
    a = temp
  END WHILE
  RETURN a
END FUNCTION
```

# Question 6: Symbols Used in Flowcharts

Standard flowchart symbols include:

- Oval: Start/End
- Rectangle: Process
- Diamond: Decision
- Parallelogram: Input/Output
- Arrow: Flow direction
